```python
import numpy as np
import pandas as pd
import csv
```

```python
food_inspect = pd.read_csv("Food_Inspections.csv")
```

```python
#what's the size of the df? 
len(food_inspect)
```

```python
food_inspect.head()
```

```python
#convert inspection date to datetime object 
#year-month-day
food_inspect['Inspection Date'] = pd.to_datetime(food_inspect['Inspection Date'])
food_inspect.head()
```

```python
#all inspections since Jan 1, 2019
before_covid = food_inspect[food_inspect['Inspection Date'] > '2019-01-01']
```

```python
#cut down so from jan 1, 2019 - march 1, 2020 
before_covid = before_covid[before_covid["Inspection Date"] < "2020-03-01"]
len(before_covid)
```

```python
#sort it by date
before_covid = before_covid.sort_values(by = ["Inspection Date"])
```

```python
before_covid.head()
```

```python
before_covid.tail()
```

```python
#inspect_subset.to_csv("inspect_rests.csv")
#still too big to upload to github. Maybe cutting down the columns will help? 
```

```python
before_covid = before_covid.drop(columns = ["Inspection ID", "License #", "Risk", "City", "State"])
```

```python
before_covid.head()
```

```python
#count number of duplicated restaurants, think I'll drop these but want to check first 
duplicated_names = before_covid.duplicated(subset=['DBA Name'])
```

```python
duplicated_names.value_counts()
#lots of duplicates
```

```python
#since these are sorted by inspection date, I can drop the duplicates and I'll be dropping the older "out of date" values
#should be ending up with 10,740 unique food establishments prior to covid 
#want to keep the first occurence in the list as that's the most recent status (still before covid)
before_covid = before_covid.drop_duplicates('DBA Name', keep='first')
```

```python
len(before_covid)
```

```python
#now let's see which ones are still open at the most recent version of pre-covid inspections 
before_covid.Results.value_counts()
```

```python
before_covid = before_covid[before_covid.Results != "Out of Business"]
```

```python
len(before_covid)
```

```python
#save the pre-covid dataset with all the still open businesses in 2019 - March 2020 period 
before_covid.to_csv("before_covid.csv")
```

```python
###
#now to the restaurants after covid 
#all restaurants inspected after 3/1/2020 (in this format its year-month-day)
after_covid = food_inspect[food_inspect['Inspection Date'] > '2020-03-01']
```

```python
len(after_covid)
```

```python
after_covid.head()
```

```python
#sort by date 
after_covid = after_covid.sort_values(by = ["Inspection Date"])
#drop extra columns 
after_covid = after_covid.drop(columns = ["Inspection ID", "License #", "Risk", "City", "State"])
```

```python
#how many duplicates in this dataset? 
```

```python
after_duplicated_names = after_covid.duplicated(subset=['DBA Name'])
after_duplicated_names.value_counts()
#this is going to make the dataset much smaller than the pre-covid one 
#makes sense as there's likely fewer resources to fully inspect as they had the previous year 
```

```python
#keep most recent inspection results 
after_covid = after_covid.drop_duplicates('DBA Name', keep='first')
```

```python
len(after_covid)
```

```python
#of these how many are closed? 
after_covid.Results.value_counts()
#1,039 are now out of business 
```

```python
#save the dataset with all food places inspected since 3/1/2020 (includes open AND closed businesses)
after_covid.to_csv("after_covid.csv")
```

```python
#now let's specifically look at the ones that are closed and see how that compares to their status last year 
now_closed = after_covid[after_covid.Results == "Out of Business"]
```

```python
len(now_closed)
```

```python
#save the restaurants that have been marked as closed within the past year 
now_closed.to_csv("now_closed.csv")
```

```python
#merge this with the before data set to get just the closed restaurants and their prior status
#I think this means a left merge
merged_closed = now_closed.merge(before_covid, how='left', on= "DBA Name", suffixes = ("_after", "_before"))
```

```python
len(merged_closed)
```

```python
merged_closed.head()
```

```python
#were all the now closed restaurants included in the before data set? 
merged_closed.count()
#so about half of these restaurants weren't included in the original data set? 
```

```python
merged_closed.Results_before.value_counts()
#good so none were closed before the pandemic 
```

```python
#what happens when we try merging the results of the before and after covid data sets? - all of the restaurants 
#try to identify both the ones that are actively closed and ones that weren't/couldn't be inspected in 2020 (might be closed?)
#perserve all values = outer join
merged_all = before_covid.merge(after_covid, how='outer', on= "DBA Name", suffixes = ("_before", "_after"))
```

```python
len(merged_all)
```

```python
merged_all.count()
#so there are fewer restaurants examined in 2020
#around 1,936 fewer restaurants examined and this doesn't include missings!
#let's search for these restaurants too? 
```

```python
merged_all.to_csv("merged_all.csv")
```

```python
#select those restaurants that are NaN for results after but had information in 2019 
missing_after = merged_all[merged_all['Results_after'].isna()]
```
